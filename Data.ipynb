# Creating First Big Data

```python
def calculate_features(Path = 'C:\\Users\\ali\\finance_Data\\Data',t=1):
    namad = os.listdir(Path)
    Origin_Key = ['TICKER','date','open','high','low','close','volume','value','num','YES-PRICE','code','EN-NAME','NAME','SHAMSI','akharin']
    key = ['code', 'date', 'open', 'high', 'low', 'close', 'volume', 'value', 'num', 'akharin']
    total = pd.DataFrame()
    for i in namad:
        data = pd.read_excel(Path+'\\'+i)
        data.columns = Origin_Key
        data = data[key]
        
        data = data_generator(data,t)
        total = pd.concat([total, data], ignore_index=True)

    data = total.dropna(axis=0)
    
    name = 'all_'+str(t)+'.csv'
    data.to_csv(name, index=False)
    print ('Done!')
```

# creating Info sheet (namad list and start & end date)

```python
def create_namad_list(data,Path = 'C:\\Users\\ali\\finance_Data\\Data',t)
    namad = os.listdir(Path)
    name = []
    path = []
    start = []
    end = []
    exist = []
    for i in namad:
        exist = np.sign(data.COD.str.contains(i.split('.')[0]).sum())
        if exist == 0:
            name.append(i.split('.')[0])
            path.append(Path + '\\' + i)
            start.append(0)
            end.append(0)

    X = {'name':name , 'path':path  , 'start':start , 'end':end }
    Info = pd.DataFrame(X)
    
    name = 'Info_'+str(t)+'.csv'
    Info.to_csv(name, index=False)
    print('Done!')
```
